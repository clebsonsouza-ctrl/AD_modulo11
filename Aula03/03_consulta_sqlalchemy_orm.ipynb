```python
from sqlalchemy import create_engine, URL, Column, String, Integer, Boolean
from sqlalchemy.orm import sessionmaker, declarative_base
from config import DB_CONFIG
import pandas as pd
```

```python
# 01_conexão com o banco aula03

url_conexao = URL.create(
    drivername="postgresql+psycopg2",
    username=DB_CONFIG["user"],
    password=DB_CONFIG["password"],
    host=DB_CONFIG["host"],
    port=DB_CONFIG["port"],
    database="aula03"
)

engine = create_engine(url_conexao)
```

```python
#02_criar sessão

Session = sessionmaker(bind=engine)
session = Session()

Base = declarative_base()
```

```python
#03_Mapear a tabela usuários

class Usuario(Base):
    __tablename__ = "usuarios"

    id = Column(Integer, primary_key=True, autoincrement=True)
    nome = Column(String(100))
    email = Column(String(255))
    senha = Column(String(100))
    ativo = Column(Boolean)
```

```python
#04_Consultar os dados do usuário

#faz uma consulta ORM na tabela representada pela classe Usuario
usuarios = session.query(Usuario).all()
```

```python
# 05_exibir usuários

for usuario in usuarios:
       print(usuario.id, usuario.nome, usuario.email, usuario.ativo)
```

```python
#06_Exibir em um dataframe

dados_usuarios = []

for usuario in usuarios:
    dados_usuarios.append({
        "id": usuario.id,
        "nome": usuario.nome,
        "email": usuario.email,
        "senha": usuario.senha,
        "ativo": usuario.ativo
    })

df_usuarios = pd.DataFrame(dados_usuarios)

df_usuarios
```

```python
# 07_Encerrar sessão e engine

session.close()
engine.dispose()

print("Consulta finalizada com sucesso.")
```
