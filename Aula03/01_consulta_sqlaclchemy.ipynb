```python
# pip install sqlalchemy

import pandas as pd
from sqlalchemy import create_engine, URL
from config import DB_CONFIG

#Isso não abre conexão com o banco. Apenas monta a configuração.
#dialect+driver://username:password@host:port/database
url_conexao = URL.create(
    drivername="postgresql+psycopg2", #isso é novo
    username=DB_CONFIG["user"],
    password=DB_CONFIG["password"],
    host=DB_CONFIG["host"],
    port=DB_CONFIG["port"],
    database="aula01"
)

# CRIAR ENGINE DO SQLALCHEMY
#gerenciar a comunicação entre o Python e o banco de dados
engine = create_engine(url_conexao)

# CONSULTA SQL
query = """
SELECT * FROM vw_resumo_vendas;
"""

with engine.connect() as conexao:
    df_resumo_vendas = pd.read_sql_query(query, conexao)

# opcional
engine.dispose()


# EXIBIR RESULTADO
display(df_resumo_vendas)

print("Consulta realizada com SQLAlchemy e pandas.")
```
