```python
# Importar bibliotecas

#!pip install psycopg2-binary
import pandas as pd
import psycopg2
from config import DB_CONFIG
```

```python
# 01 Conecetar ao banco postgres

conexao = psycopg2.connect(
    host=DB_CONFIG["host"],
    database=DB_CONFIG["database"],
    user=DB_CONFIG["user"],
    password=DB_CONFIG["password"],
    port=DB_CONFIG["port"]
)

conexao.autocommit = True

cursor = conexao.cursor()
```

```python
#02 Criar banco caso não exista

cursor.execute("SELECT 1 FROM pg_database WHERE datname = 'aula02';")
banco_existe = cursor.fetchone()

if banco_existe:
    print("O aula02 já existe.")
else:
    cursor.execute("CREATE DATABASE aula02;")
    print("Banco aula02 criado com sucesso.")

cursor.close()
conexao.close()

print("Conexão com o banco postgres encerrada.")
```

```python
# 03 Conexão com o banco aula02

conexao = psycopg2.connect(
    host=DB_CONFIG["host"],
    database="aula02", #alterado para o banco criado
    user=DB_CONFIG["user"],
    password=DB_CONFIG["password"],
    port=DB_CONFIG["port"]
)

cursor = conexao.cursor()

print("Conexão com o banco aula02.")
```

```python
# 04 Criar a tabela

query_criar_tabela = """
CREATE TABLE IF NOT EXISTS alunos (
    id SERIAL PRIMARY KEY,
    nome VARCHAR(100) NOT NULL,
    idade INTEGER,
    cidade VARCHAR(100),
    data_cadastro TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
"""

cursor.execute(query_criar_tabela)
conexao.commit()

print("Tabela alunos criada com sucesso.")
```

```python
# 05 Inserir dados na tabela

query_inserir_dados = """
INSERT INTO alunos (nome, idade, cidade)
VALUES
    ('Ana', 25, 'João Pessoa'),
    ('Carlos', 30, 'Recife');
"""

cursor.execute(query_inserir_dados)
conexao.commit()

print("Dados inseridos com sucesso.")
```

```python
# 06 Consultar tabela exibir dataframe

query_consultar_dados = """
SELECT
    id, nome, idade, cidade, data_cadastro
FROM alunos
ORDER BY id;
"""

df_alunos = pd.read_sql_query(query_consultar_dados, conexao)

display(df_alunos)
```

```python
# 07 Fechar cursor e conexao

if cursor:
    cursor.close()
    print("Cursor encerrado com sucesso.")

if conexao:
    conexao.close()
    print("Conexão com o banco02 encerrada com sucesso.")
```
