EXTRACT (EXTRAIR)

Exemplo desemprego

```python
# Indicador Banco Mundial / ILO:
# SL.UEM.TOTL.ZS = Unemployment, total (% of total labor force) (modeled ILO estimate)

indicador = 'SL.UEM.TOTL.ZS'

url = f"https://api.worldbank.org/v2/en/indicator/{indicador}?downloadformat=csv"

response = requests.get(url)
response.raise_for_status()


zip_file = ZipFile(BytesIO(response.content))

arquivo_dados = [
    nome for nome in zip_file.namelist()
    if nome.startswith(f"API_{indicador}") and nome.endswith(".csv")
][0]

df = pd.read_csv(
    zip_file.open(arquivo_dados),
    skiprows=4
)

df
```

TRANSFORM (Transformar)

```python
# FILTRANDO PELAS AS 15 MAIORES ECONOMIA

paises = {
    "United States": "Estados Unidos",
    "China": "China",
    "Germany": "Alemanha",
    "Japan": "Japão",
    "India": "Índia",
    "United Kingdom": "Reino Unido",
    "France": "França",
    "Italy": "Itália",
    "Brazil": "Brasil",
    "Canada": "Canadá",
    "Russian Federation": "Rússia",
    "Mexico": "México",
    "Korea, Rep.": "Coreia do Sul",
    "Australia": "Austrália",
    "Spain": "Espanha"
}

df_15 = df.copy()

df_15 = df_15[df_15["Country Name"].isin(paises.keys())].copy()

df_15["pais"] = df_15["Country Name"].map(paises)
```

```python
df_15
```

```python
anos = [str(ano) for ano in range(1991, 2026)]

df_desem = df_15.melt( 
    id_vars=["pais", "Country Code"], 
    value_vars=anos, 
    var_name="ano", 
    value_name="desemprego" 
)

df_desem["ano"] = df_desem["ano"].astype(int)

df_desem["desemprego"] = pd.to_numeric(df_desem["desemprego"], errors="coerce")

df_desem = df_desem.dropna(subset=["desemprego"])

df_desem = df_desem.sort_values(["ano", "desemprego"], ascending=[True, False])

print(df_desem.info())

df_desem
```

LOAD (Carregar)

```python
#conectar como banco aula04

url_conexao = URL.create(
    drivername="postgresql+psycopg2",
    username=DB_CONFIG["user"],
    password=DB_CONFIG["password"],
    host=DB_CONFIG["host"],
    port=DB_CONFIG["port"],
    database="aula04" # alterando para aula04    
)

engine = create_engine(url_conexao)

print("Conexão com o aula04")
```

```python
df_desem.to_sql(
    name="desemprego_15",
    con=engine,
    if_exists="replace", #"append"
    index=False,
    dtype={
        "pais": String(),
        "Country Code": String(),
        "ano": Integer(),
        "desemprego": Float()
    }
)

print('Tabela criada/carregada com sucesso!')

# fecha/libera as conexões do SQLAlchemy
engine.dispose()
```
